```python
import sounddevice as sd
import numpy as np
import wave
import threading
import sys
import tempfile
import whisper
import os
import subprocess
# Parameters
sample_rate = 44100  # 44.1 kHz
channels = 1  # 1 = Mono, 2 = Stereo
file_name = "output_sounddevice.wav"
dtype = 'int16'

# Buffer to hold recorded data
recorded_frames = []
recording = True
model = whisper.load_model("tiny.en")
```

```python
# def transcribe_to_txt(input_filename: str, output_filename: str):
#     print('Running whisper transcription...')
#     # temp = tempfile.NamedTemporaryFile(dir=".").name
#     f = open(input_filename, 'w')
#     print(f)

#     # Transcribe the audio file
#     print("begin transcription")
#     result = model.transcribe(input_filename)
#     print("end transcription")
#     # Save to a text file
#     with open(output_filename, "w", encoding="utf-8") as f:
#         f.write(result["text"])
# def transcribe_to_txt(input_filename: str, output_filename: str):
#     print('Running whisper transcription...')
#     # Compose the command of all components
#     command = ['./main', '-f', input_filename, '-otxt', '-of', output_filename, '-np']
#     # Execute the command
#     result = subprocess.run(command, capture_output=True, text=True)

def callback(indata, frames, time, status):
    # Raise for status if required
    if status:
        print(status)
    tmpfile_name = None
    # Create a tempfile to save the audio to, with autodeletion
    with tempfile.NamedTemporaryFile(delete=False, suffix='.mp3', prefix='audio_',dir = ".") as tmpfile:
        # Save the 5 second audio to a .wav file
        wave_file = wave.open(tmpfile,'wb')
        wave_file.setnchannels(1)  # Mono audio
        wave_file.setsampwidth(2)  # 16-bit audio
        wave_file.setframerate(16000)  # Sample rate
        wave_file.writeframes(indata)
        wave_file.close()
        tmpfile_name = tmpfile.name
        # Prepare the output filename
    print(tmpfile_name)
    output_filename = tmpfile_name.replace('.mp3', '')
    result = model.transcribe(tmpfile_name)
    # result = transcribe_to_txt(tmpfile.name, output_filename)
    print("finish transcribing")
    with open(output_filename + '.txt', "w", encoding="utf-8") as f:
        f.write(result["text"])

    # Print the transcribed text
    with open(output_filename + '.txt', 'r') as file:
        print("reading output")
        print(file.read())
    
    # Clean up temporary files
    os.remove(output_filename + '.txt')
    os.remove(tmpfile.name)
```

```python
try:
    # Start recording with a rolling 5-second buffer
    with sd.InputStream(callback=callback, dtype='int16', channels=1, samplerate=16000, blocksize=16000*5):
        print("Recording... Press Ctrl+C to stop.")
        while True:
            pass
except KeyboardInterrupt:
    print('Recording stopped.')
```
